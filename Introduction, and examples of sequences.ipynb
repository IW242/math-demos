### Warmup example

Let's code up our first sequence in Python, say the sequence $a_n = \frac{\sin n}{\sqrt{n}}$.

```python
def a(n):
    return np.sin(n)/np.sqrt(n)
```

Ok, let's see how that works. Let's evaluate a few terms.

```python
print("a_1 = " , a(1), ",", "a_2 = ", a(2), ",", "a_3 = ", a(3))
```

Let's write it down in a pretty table. (Does anyone know how to output this even prettier?)

```python
for n in range(1,11):
    print("| \t a_%d \t | \t %s \t |" % (n, format(a(n), ".7f")))
```

Now let's plot that as a graph.

```python
N = 50 # The number of terms we'll plot in the sequence
plt.xkcd() # I love this XKCD style!
n = np.linspace(1, N, N) # The array of N numbers 1, ..., N.
ax = pylab.axes()
ax.plot(n, a(n), 'o')
ax.set_title('a_n = n/n+1')
ax.set_xlabel('n')
ax.set_ylabel('a_n')
```

It's kind of complicated because the points are all over the place! It's easier if we also plot the underlying function $f(x) = \frac{\sin x}{\sqrt{x}}$:

```python
N = 50 # The number of terms we'll plot in the sequence
plt.xkcd() # I love this XKCD style!
n = np.linspace(1, N, N) # The array of N numbers 1, ..., N.
x = np.linspace(1, N, 100) # 100 points, much more.
ax = pylab.axes()
ax.plot(x, a(x), color="red", label="f(x)")
ax.plot(n, a(n), "o", color="blue", label="a_n")
ax.set_title('a_n = n/n+1')
ax.set_xlabel('n')
ax.set_ylabel('a_n')
ax.legend(loc="upper right")
```

The sequence seems to be converging to 0. Let's check some more entries in the table.

```python
for n in range(1000,1005):
    print("| \t a_%d \t | \t %s \t |" % (n, format(a(n), ".7f")))
```

Yes... looks like it's converging to zero.

### The sequence giving Euler's number e

Ok, now we're ready to try a more tricky sequence, the famous Euler sequence which converges to the number $e=2.781828...$:

$$ e = \lim_{n\rightarrow \infty} (1 + \frac{1}{n})^n $$

We code this sequence into Python as before:

```python
def a(n):
    return (1+1/n)**n 
```

Print out the first few terms as a table:

```python
for n in range(1,11):
    print("| a_%d \t | %s |" % (n, format(a(n), ".7f")))
```

Seems to be converging quite slowly! Let's plot that as a graph.

```python
N = 25  
n = np.linspace(1, N, N) 
ax = pylab.axes()
ax.plot(n, a(n), 'o', color="red")
ax.set_title("a_n = (1+1/n)^n")
ax.set_xlabel('n')
ax.set_ylabel('a_n')
```

Ok, looks to be converging! 
Finally, let's compare the exact value of $e$ correct to 15 decimal places, with, say, the 100th term of our sequence:

```python
print("The exact value of e, to 15 decimal places, is ", np.e)
print("The 100th term in the sequence is", a(100))
print("The 1000th term in the sequence is", a(1000))
```

So, the convergence is relatively slow! After 1000 terms we're not even correct to 2 decimal places!

## A sequence for $\pi$

Now let's write down a sequence converging to $pi$, using the formula (a consequence of the Prime Number Theorem):
$$ {\displaystyle \pi = lim_{n \rightarrow \infty} n^{\frac{\text{primes}(n)}{n}}} $$
Here $\mbox{primes}(n)$ is the function which counts the number of prime numbers less than or equal to $n$. So we'll have to code that first.

```python
def isprime(n):
    '''check if integer n is a prime'''
    # make sure n is a positive integer
    n = abs(int(n))
    # 0 and 1 are not primes
    if n < 2:
        return False
    # 2 is the only even prime number
    if n == 2: 
        return True    
    # all other even numbers are not primes
    if not n & 1: 
        return False
    # range starts with 3 and only needs to go up the squareroot of n
    # for all odd numbers
    for x in range(3, int(n**0.5)+1, 2):
        if n % x == 0:
            return False
    return True

def number_of_primes_less_than(n):
    count = 0
    for i in range(1, n+1):
        if isprime(i):
            count = count + 1
    return count
```

```python
def a(n):
    return n**(number_of_primes_less_than(n)/n) 
```

```python
for n in range(1,20):
    print("| a_%d \t | %s |" % (n, format(a(n), ".7f")))
```

Now we plot the graph!

```python
# First make the array of x and y values
N = 50
x_array = []
y_array = []
for n in range(1,N+1):
    x_array.append(n)
    y_array.append(a(n))

# Now plot the graph.
ax = pylab.axes()
ax.plot(x_array, y_array, color="blue")
ax.plot(x_array, y_array, 'o', color="red")
ax.set_xlabel('n')
ax.set_ylabel('a_n')
```

Seems to be converging to $\pi$!

Let's check - we'll compute $a_n$ when $n$ is one million!

The convergence of the expession $ {\displaystyle \pi = lim_{n \rightarrow \infty} n^{\frac{\text{primes}(n)}{n}}} $ is very slow. In fact, the rate at which it converges is a famous unsolved problem in mathematics, called the [Riemann hypothesis](https://en.wikipedia.org/wiki/Riemann_hypothesis)!

But, the [Prime Number Theorem](https://en.wikipedia.org/wiki/Prime_number_theorem) says that the limit is indeed $\pi$. It's just no one is sure exactly how fast it converges to $\pi$.
